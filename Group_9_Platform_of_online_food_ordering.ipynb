Pie chart to display Food ordering online based on Occupation Range

```python
df_var = df_rank.groupby('Occupation').size()
df_var

df_var.plot(kind='pie',
                            figsize=(10, 6),
                            autopct='%1.1f%%',
                            startangle=0,
                            shadow=False,
                            labels=None,         # turn off labels on pie chart
                            pctdistance=1.12,    # the ratio between the center of each pie slice and the start of the text generated by autopct
                            )
plt.title('Food ordering online based on Occupation Range', y=1.12, fontsize = 15)
plt.axis('equal')

# add legend
plt.legend(labels=df_var.index, loc='upper left', fontsize=7)

plt.show()
```

This pie chart give the categorial information about people who are at the different occupation. It is clear from the graph that mostly students are involved in ordereing food online which is more than half of the total values.

***Visualizing data using columns Latitude and longitude from the given data set to figure out the area where the online orders were made from***


```python
import folium
```

**Displaying Area map of Bangalore using folium**


```python
area_map = folium.Map(location=[12.9716, 77.5946], zoom_start=11.5)


area_map
```

```python
incidents = folium.map.FeatureGroup()

# loop through the 100 crimes and add each to the incidents feature group
for lat, lng, in zip(df_rank.latitude, df_rank.longitude):
    incidents.add_child(
        folium.vector_layers.CircleMarker(
            [lat, lng],
            radius=5, # define how big you want the circle markers to be
            color='red',
            fill=True,
            fill_color='yellow',
            fill_opacity=0.6
        )
    )

# add incidents to map
area_map.add_child(incidents)
```

```python
age_group = df_rank.groupby(['Age']).size()
x = age_group.index
y= age_group.values
plt.bar(  x,  y )
plt.title('Number of Order Age wise')
plt.xlabel('Age')
plt.ylabel('Number of Orders')
plt.show()
```

```python
# Bar Chart using plotly
import plotly.graph_objects as go
```

```python
colors = ['blue' if val < 35 else 'red' for val in y]
trace = go.Bar(x= x , y = y , marker=dict(color=colors))
fig = go.Figure(trace)
fig.update_layout( title='Number of Orders Age wise',     xaxis=dict(title='Age'),     yaxis=dict(title='Number of orders') )
fig.show()
```

```python
income_group = df_rank.groupby(['Monthly Income']).size()
x = income_group.index
y= income_group.values
trace = go.Bar(x= x , y = y )
fig = go.Figure(trace)
fig.update_layout( title='Number of Order Monthly Income wise',     xaxis=dict(title='Monthly Income'),     yaxis=dict(title='Number of orders') )
fig.show()
```
